# D1 = D2 = 25 mm

```python
mm_25_data = pd.read_excel(filepath, sheet_name='Diameter_25mm')
mm_25_data.head()
```

```python
required_data = mm_25_data[['Gap Width Experimental (in mm)', 'Offset from Centre', "W'", 'Bubble Behaviour']]
required_data = required_data.dropna(how='all')
required_data = required_data.fillna(method='ffill')
required_data["b"] = required_data["Gap Width Experimental (in mm)"]*required_data['Offset from Centre']
required_data["b'"] = 2 * required_data["b"]/required_data["Gap Width Experimental (in mm)"]
required_data
```

```python
required_data['Bubble Behaviour'].unique()
```

```python
mm_25_spherical_collapse = required_data[required_data['Bubble Behaviour'] == 'Spherical Collapse']
mm_25_jetting_towards = required_data[required_data['Bubble Behaviour'] == 'Jetting Towards Nearer Wall']
mm_25_jetting_away = required_data[required_data['Bubble Behaviour'] == 'Jetting Away From Nearer Wall']
mm_25_split_collapse = required_data[required_data['Bubble Behaviour'] == 'Split Collapse']
mm_25_centre_collapse = required_data[required_data['Bubble Behaviour'] == 'Centre Collapse']
```

```python
plt.scatter(mm_25_jetting_towards["W'"], mm_25_jetting_towards["b"], color = 'blue', label = "Jetting Towards Nearer Wall")
plt.scatter(mm_25_spherical_collapse["W'"], mm_25_spherical_collapse["b"], color = 'red', label = "Spherical Collapse")
plt.scatter(mm_25_split_collapse["W'"], mm_25_split_collapse["b"], color = 'fuchsia', label = "Split Collapse")
plt.scatter(mm_25_jetting_away["W'"], mm_25_jetting_away["b"], color = 'navy', label = "Jetting Away From Nearer Wall")
plt.scatter(mm_25_centre_collapse["W'"], mm_25_centre_collapse["b"], color = 'olive', label = "Centre Collapse")
plt.xlabel("W'", fontweight = "bold"), plt.ylabel("b", fontweight = "bold")
plt.legend(fontsize = 9)
plt.title('Bubble Behaviour, D1 = D2 = 25 mm')
plt.show()
#plt.savefig("/content/drive/MyDrive/IITH SURE/Plots/D1 = D2 = 25 mm.jpg")
```

```python
plt.scatter(mm_25_jetting_towards["b'"], mm_25_jetting_towards["W'"], color = 'blue', label = "Jetting Towards Nearer Wall")
plt.scatter(mm_25_spherical_collapse["b'"], mm_25_spherical_collapse["W'"], color = 'red', label = "Spherical Collapse")
plt.scatter(mm_25_split_collapse["b'"], mm_25_split_collapse["W'"], color = 'fuchsia', label = "Split Collapse")
plt.scatter(mm_25_jetting_away["b'"], mm_25_jetting_away["W'"], color = 'navy', label = "Jetting Away From Nearer Wall")
plt.scatter(mm_25_centre_collapse["b'"], mm_25_centre_collapse["W'"], color = 'olive', label = "Centre Collapse")
plt.xlabel("b'", fontweight = "bold"), plt.ylabel("W'", fontweight = "bold")
plt.legend(loc='upper left', bbox_to_anchor=(1.001, 1), fontsize = 9)
plt.title('Bubble Behaviour, D1 = D2 = 25 mm')
plt.show()
```
